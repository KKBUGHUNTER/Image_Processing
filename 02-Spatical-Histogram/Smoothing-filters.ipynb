```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the color image
image = cv2.imread('nature.jpg')
```

# Smoothing Filters: Standard Average and Weighted Average Filters on a Color Image

```python


# Convert the image to RGB (OpenCV reads images in BGR format by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define kernels for filters
average_kernel = np.ones((3, 3), np.float32) / 9  # 3x3 averaging filter kernel
weighted_avg_kernel = np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]], np.float32) / 16  # 3x3 weighted averaging filter kernel

# Split the color image into individual channels (R, G, B)
channel_r, channel_g, channel_b = cv2.split(image_rgb)

# Apply smoothing filters to each channel separately
smoothed_r = cv2.filter2D(channel_r, -1, average_kernel)
smoothed_g = cv2.filter2D(channel_g, -1, average_kernel)
smoothed_b = cv2.filter2D(channel_b, -1, average_kernel)

weighted_smoothed_r = cv2.filter2D(channel_r, -1, weighted_avg_kernel)
weighted_smoothed_g = cv2.filter2D(channel_g, -1, weighted_avg_kernel)
weighted_smoothed_b = cv2.filter2D(channel_b, -1, weighted_avg_kernel)

# Merge the filtered channels back into the color image
smoothed_image = cv2.merge((smoothed_r, smoothed_g, smoothed_b))
weighted_smoothed_image = cv2.merge((weighted_smoothed_r, weighted_smoothed_g, weighted_smoothed_b))

# Display the original and filtered images using Matplotlib
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(smoothed_image)
plt.title('Standard Average Filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(weighted_smoothed_image)
plt.title('Weighted Average Filter')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Order Statistics Filters and Gaussian Filter

```python


# Apply order statistics filters
median_filtered = cv2.medianBlur(image, 3)  # Median filter with a 3x3 kernel
max_filtered = cv2.dilate(image, np.ones((3, 3), np.uint8))  # Max filter using dilation
min_filtered = cv2.erode(image, np.ones((3, 3), np.uint8))  # Min filter using erosion

# Apply Gaussian filter
gaussian_filtered = cv2.GaussianBlur(image, (3, 3), 0)  # Gaussian filter with a 3x3 kernel

# Display the original and filtered images using Matplotlib
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filter')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(max_filtered, cmap='gray')
plt.title('Max Filter')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(min_filtered, cmap='gray')
plt.title('Min Filter')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(gaussian_filtered, cmap='gray')
plt.title('Gaussian Filter')
plt.axis('off')

plt.tight_layout()
plt.show()
```
