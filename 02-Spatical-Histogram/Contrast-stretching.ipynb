```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the image
image = cv2.imread('images.jpeg')


```

# Invert the image


```python
# Invert the image
negative_image = 255 - image

# Display original and negative images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))
plt.title('Negative Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Log transformation


```python


c = 1  # Constant for scaling
log_transformed = c * np.log1p(image)

# Normalize the log-transformed image to 0-255 range
log_transformed = (255 * log_transformed / np.max(log_transformed)).astype(np.uint8)

# Display original and log-transformed images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(log_transformed, cmap='gray')
plt.title('Log Transformed Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Gamma transformation


```python
gamma = 0.5  # Gamma value (adjust as needed)
gamma_transformed = np.power(image / 255.0, gamma)
gamma_transformed = (255 * gamma_transformed).astype(np.uint8)

# Display original and gamma-transformed images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gamma_transformed, cmap='gray')
plt.title('Gamma Transformed Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Perform contrast stretching


```python

# Calculate minimum and maximum pixel values
min_val = np.min(image)
max_val = np.max(image)

# Define the desired minimum and maximum intensity values for stretching
new_min = 0  # Adjust this value for desired minimum intensity
new_max = 255  # Adjust this value for desired maximum intensity

# Perform contrast stretching
stretched_image = ((image - min_val) * (new_max - new_min) / (max_val - min_val)) + new_min
stretched_image = stretched_image.astype(np.uint8)

# Display original and contrast-stretched images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(stretched_image, cmap='gray')
plt.title('Contrast-Stretched Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Apply histogram equalization


```python
image = cv2.imread('images.jpeg', cv2.IMREAD_GRAYSCALE)

# Calculate and plot the histogram before equalization
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(image.flatten(), bins=256, range=[0,256], color='r')
plt.title('Histogram Before Equalization')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Calculate and plot the histogram after equalization
plt.subplot(1, 2, 2)
plt.hist(equalized_image.flatten(), bins=256, range=[0,256], color='g')
plt.title('Histogram After Equalization')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()

# Display original and equalized images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
