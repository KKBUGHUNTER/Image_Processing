# Correlation and Convolution with OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Read the image in grayscale
image = cv2.imread('images.jpeg', cv2.IMREAD_GRAYSCALE)

# Define a 3x3 averaging filter kernel for spatial filtering
kernel = np.ones((3, 3), np.float32) / 9

# Perform correlation
correlation_result = cv2.filter2D(image, -1, kernel, borderType=cv2.BORDER_CONSTANT)

# Perform convolution (equivalent to correlation with a flipped kernel)
convolution_result = cv2.filter2D(image, -1, cv2.flip(kernel, -1), borderType=cv2.BORDER_CONSTANT)

# Display the original image and results after correlation and convolution
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(correlation_result, cmap='gray')
plt.title('Correlation Result')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(convolution_result, cmap='gray')
plt.title('Convolution Result')
plt.axis('off')

plt.tight_layout()
plt.show()
```
