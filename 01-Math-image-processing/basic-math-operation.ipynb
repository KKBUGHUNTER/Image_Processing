```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Basic Arithmetic Operations on Images (Addition, Subtraction, Multiplication, Division)

```python

image1 = cv2.imread('Grayscale_Image.jpg')
image2 = cv2.imread('image2.png')

image2_resized = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Perform basic arithmetic operations
added_image = cv2.add(image1, image2_resized)
subtracted_image = cv2.subtract(image1, image2_resized)
multiplied_image = cv2.multiply(image1, image2_resized)
divided_image = cv2.divide(image1, image2_resized)

# Display the results using Matplotlib
titles = ['Added Image', 'Subtracted Image', 'Multiplied Image', 'Divided Image']
images = [added_image, subtracted_image, multiplied_image, divided_image]

plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Logical operations like NOT, OR, AND, and XOR on images

```python
image1 = cv2.imread('Grayscale_Image.jpg')
image2 = cv2.imread('image2.png')

image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Perform basic logical operations
not_image = cv2.bitwise_not(image1_gray)
or_image = cv2.bitwise_or(image1_gray, image2_gray)
and_image = cv2.bitwise_and(image1_gray, image2_gray)
xor_image = cv2.bitwise_xor(image1_gray, image2_gray)

# Display the results using Matplotlib
titles = ['NOT Image', 'OR Image', 'AND Image', 'XOR Image']
images = [not_image, or_image, and_image, xor_image]

plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Read an image
image = cv2.imread('Grayscale_Image.jpg', cv2.IMREAD_GRAYSCALE)  # Assuming it's a grayscale image

if image is None:
    print("Could not read the image. Please check the file path.")
else:
    # Apply mathematical operations
    LOGc = np.log1p(image.astype(np.float32))  # Convert to float for log operations
    EXPc = np.expm1(image.astype(np.float32))  # Convert to float for exp operations
    SQRTc = np.sqrt(image.astype(np.float32))  # Convert to float for sqrt operations
    TRIGc_sin = np.sin(image.astype(np.float32))  # Convert to float for trigonometric operations
    TRIGc_cos = np.cos(image.astype(np.float32))  # Convert to float for trigonometric operations
    TRIGc_tan = np.tan(image.astype(np.float32))  # Convert to float for trigonometric operations
    INVERTc = (255 - image)  # Inversion

    # Display the results using Matplotlib
    titles = ['LOGc', 'EXPc', 'SQRTc', 'TRIGc_sin', 'TRIGc_cos', 'TRIGc_tan', 'INVERTc']
    images = [LOGc, EXPc, SQRTc, TRIGc_sin, TRIGc_cos, TRIGc_tan, INVERTc]

    plt.figure(figsize=(12, 10))
    for i in range(len(images)):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```
