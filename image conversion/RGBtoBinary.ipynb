## load image using CV2

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cv2 

image = cv2.imread('image.png')
```

## Convert Blue-Green-Red image to GrayScale image.

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

image_matrix = np.array(gray_image)

print(image_matrix)
```

## Find the Image Size using PIL

```python
pilimage = Image.open('image.png')

width, height = pilimage.size

print("Size of the image: ", width, " X ", height)
```

## Display the image using matplotlib.pyplot

```python
plt.imshow(image)

plt.show()
```

## Convert the GrayScale image to Black and White image. (Manually)

```python
limit = 125

black_and_white_image_matrix = gray_image.copy()

for i in range(len(gray_image)):
    for j in range(len(gray_image[i])):
        if(gray_image[i][j] <= limit):
            black_and_white_image_matrix[i][j] = 0
        else:
            black_and_white_image_matrix[i][j] = 1

print(black_and_white_image_matrix)

bawwidth, bawheight = len(black_and_white_image_matrix),len(black_and_white_image_matrix[0])

print("Size of the newly created black and white image: ", bawwidth, " X ", bawheight)

plt.imshow(black_and_white_image_matrix)
plt.show()
```

```python
for i in black_and_white_image_matrix:
    print(i)
if(cv2.imwrite("black_and_white_image.jpg",black_and_white_image_matrix)):
    print("Image Saved Successfuly..")
```
